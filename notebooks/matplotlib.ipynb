## Importando Bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gs
```

## Funções Utilizadas

```python
def normal(x):
    return np.exp(-(x**2))
def graph_3d(X,Y):
    return np.exp(-(X**2+Y**2))+2*np.exp(-((X+1.7)**2+(Y+1.7)**2))
def exponencial(x):
    return np.exp(-x)
def seno(x):
    return np.sin(x*2*np.pi)
def cosseno(x):
    return np.cos(x*2*np.pi)
def conv(x,y):
    return np.convolve(x,y)
```

## Figura 1: Exemplo combinado.

```python
x_axis = np.linspace(-2,2,1000)
plt.figure(0,facecolor='k') # facecolor é referente à cor externa da figura
y = seno(x_axis)*exponencial(x_axis)
plt.plot(x_axis,y)
y = cosseno(x_axis)*exponencial(x_axis)
plt.plot(x_axis,y)
plt.grid()
plt.gca().set_facecolor('k') # a facecolor da gca é referente à cor interna da figura
```

## Figura 2: Exemplo em quadrados separados.

```python
fig,ax = plt.subplots(2,2)
ax[0,0].plot(x_axis,normal(x_axis))
ax[0,0].set_facecolor('g')
x_axis = np.linspace(-2,2,50)
ax[0,1].stem(x_axis,seno(x_axis))
ax[1,0].scatter(x_axis,cosseno(x_axis),marker='1',s=25)
ax[1,1].violinplot(x_axis)
```

## Figura 3: Exemplo em separado 3 plots (GridSpec):

```python
fig = plt.figure(2)
g = gs.GridSpec(2,2)
ax1 = plt.subplot(g[0,0])
ax2 = plt.subplot(g[0,1])
ax3 = plt.subplot(g[1,:])
x = np.random.randint(0,50,size=(50,))
ax1.pie(x)
ax2.plot(x_axis,seno(x_axis),color='blue',linestyle='dashed',linewidth=2)
x = np.random.normal(30,5,size=(1000,))
ax3.violinplot(x,vert=False)
```

## Figura 4: Exemplo em separado 3 plots  e um 3d (GridSpec):

```python
fig=plt.figure(3)
g = gs.GridSpec(2,2)
ax1 = fig.add_subplot(g[0,0],projection='3d')
ax2 = plt.subplot(g[0,1])
ax3 = plt.subplot(g[1,:])
x = np.random.randint(0,50,size=(50,))
ax2.plot(x_axis,seno(x_axis),color='blue',linestyle='dashed',linewidth=2)
x = np.random.normal(30,5,size=(1000,))
ax3.violinplot(x,vert=False)
ax1.plot(x_axis,seno(x_axis)*exponencial(x_axis),cosseno(x_axis)*exponencial(x_axis))
```

## Figura 5: Plot tridimensional simples:

```python
fig = plt.figure(5)
ax = fig.add_subplot(projection='3d')
x_axis = np.linspace(-2,2,1000)
ax.plot(x_axis,
        cosseno(x_axis)*exponencial(x_axis),
        seno(x_axis)*exponencial(x_axis),linestyle='dotted',color='r')
```

## Figura 6: Plot tridimensional com surf, scatter e countour

```python
fig = plt.figure(6)
ax = fig.add_subplot(projection='3d')
x_axis = np.linspace(-4,2,1000)
X,Y = np.meshgrid(x_axis,x_axis)
yy = np.tile(x_axis.reshape(len(x_axis),1),(1,1000))
xx = yy.T

ax.plot_surface(X,Y,graph_3d(X,Y),cmap='gray',
                alpha=0.1,               
                edgecolor='k',                                
                rstride=40,
                cstride=40)
ax.scatter(0,0,graph_3d(0,0),marker='x',s=100)
c= ax.contour(X,Y,graph_3d(X,Y),offset=-0.1)
fig.colorbar(c)
```

## Figura 7: Exemplo de contorno em 2d

```python
plt.figure(7)
c = plt.contour(X,Y,graph_3d(X,Y),cmap='jet',level=40)
plt.colorbar(c)
```
